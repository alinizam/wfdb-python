## Demo Scripts on how to use the wfdb-python package

```python
import wfdb
%matplotlib inline
```

```python
# Demo 1 - Read the wfdb file, plot the last part of the signals, and print the metadata

sig, fields=wfdb.rdsamp('sampledata/a103l') 
wfdb.plotwfdb(sig[80000:,:], fields, title='Record a103l from Physionet Challenge 2015') 
fields
```

```python
# Demo 2 - Read certain channels and sections of the WFDB file and plot them. 

sig, fields=wfdb.rdsamp('sampledata/s0010_re', channels=[14, 0, 5, 10, 11], sampfrom=100, sampto=15000)
wfdb.plotwfdb(sig, fields)
```

```python
# Demo 3 - Read the WFDB signal and annotation file and plot them together.

sig, fields=wfdb.rdsamp('sampledata/100')
annsamp=wfdb.rdann('sampledata/100', 'atr')[0]
wfdb.plotwfdb(sig, fields, annsamp=annsamp, annch=[0], title='Record 100 from MIT-BIH Arrhythmia Database')
```

### Matched waveform examples

The following sections load and plots waveforms from the MIMIC matched waveform database. These waveforms have been matched to clinical data in the MIMIC Clinical database. The input records are multi-segment (made up of multiple individual WFDB records) and relatively long.

Note that these kinds of records contain segments in which certain channels are missing. <strong>matplotlib</strong> automatically zooms in on sections without Nans in individual channels but the entire durations of the signals input into <strong>plotwfdb</strong> are actually plotted. 



```python
# Demo 4 - Read the multi-segment signal and plot waveforms from the MIMIC matched waveform database. 
sig, fields=wfdb.rdsamp('sampledata/matched/s25047/s25047-2704-05-04-10-44')
wfdb.plotwfdb(sig, fields, title='s25047/s25047-2704-05-04-10-44') 
fields
```

```python
# Demo 5 - Read the multi-segment signal and plot waveforms from the MIMIC matched waveform database. 
sig, fields=wfdb.rdsamp('sampledata/matched/s00001/s00001-2896-10-10-00-31')
# Note, very long record to plot. 
wfdb.plotwfdb(sig[240000:, :], fields, title='s00001/s00001-2896-10-10-00-31') 
fields
```
